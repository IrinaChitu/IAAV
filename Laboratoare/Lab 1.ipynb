Python specific ops

```
for i in range(10): # Returneaza un obiect iterabil care enumera pana la 10
  print(i)

print('-'*10)

# Returneaza un obiect iterabil care enumera de la 1 
# pana la 10 din 3 in 3
for i in range(1, 10, 3): 
  print(i)

x = [a**2 for a in range(10)]
y = [b**3 for b in range(10)]

print('-'*10)

for idx,elem in enumerate(x): # Returneaza si indexul elementului curent
  print(idx, elem)

print('-'*10)

for x_elem,y_elem in zip(x,y): # Combina doua liste element cu element
  print(x_elem,y_elem)
```

Clase impreuna cu liste

```
class A:
  def __init__(self,x):
    self.x = x
  
  def get_x(self):
    return self.x

our_object = A(5)
print("Valoarea lui x este", our_object.get_x())

Auri = [A(3)]*5
for obiect in Auri:
  print(obiect.x, end=' ')

print()

Auri[3].x = 6

for obiect in Auri:
  print(obiect.x, end=' ', )
```

NUMPY


```
import numpy as np

# Creare obiect de tip np.ndarray
l = [[0,2,3,4],[3,5,6,7],[4,6,7,8]]
print(l)
print()
np_l = np.array(l)
print(np_l)

# Indexare conditionala
aux = np_l
aux[aux > 5] = -1
print(aux)
```

Operatii algebrice cu matrici si vectori de tip np.ndarray

```
print(l*3)
print()
print(np_l*3)
print()
print(np_l+np_l)
print()
print(np_l*np_l) # element wise
print()
print(np.dot(np_l,np_l.T)) # matrice wise
print()
print(np.dot(np_l.T,np_l)) # matrice wise
print()
print(np_l+np_l[0]) # broadcasting
```

```
x = np.random.random(size=(3,5))
print(x)
print(x.shape)
print(x.reshape(1,15).shape)
print(x.reshape(1,-1).shape)
c = np.random.choice([1,4,7])
print(c)
```

```
x = np.array([1,2,3])
print(np.repeat(x,3), end='\n\n')
print(np.repeat(x[np.newaxis, :], 3, axis=0), end='\n\n')
print(np.repeat(x[:, np.newaxis], 3, axis=1), end='\n\n')
print(np.tile(x,3), end='\n\n')
print(np.concatenate((x,x)), end='\n\n')

cat = np.concatenate((x[np.newaxis,:2],x[np.newaxis, :2]), axis=0)
print(cat, end='\n\n')
print(np.concatenate((cat, cat), axis=1))
```
